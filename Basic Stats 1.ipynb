## Bar Chart Analysis for Categorical Column:
#### Identify categorical columns in the dataset.
#### Create bar charts to visualize the frequency or count of each category.
#### Analyze the distribution of categories and provide insights.

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_columns = ['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model']    # To identify categorical columns

plt.figure(figsize=(12, 25))  # Set up the visualizations

for i, column in enumerate(categorical_columns):       # Ploting the bar charts for each categorical column
    plt.subplot(len(categorical_columns), 1, i + 1)
    sns.countplot(y=sales[column], order=sales[column].value_counts().index, color='skyblue')
    plt.title(f'Bar Chart of {column}', fontsize=14)
    plt.xlabel('Count')
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```

## Standardization of Numerical Variables

####  â—	Objective: To scale numerical variables for uniformity, improving the datasetâ€™s suitability for analytical models.
  
####  â—Steps:
	Explain the concept of standardization (z-score normalization).
	Standardize the numerical columns using the formula: z=x-mu/sigma
	Show before and after comparisons of the data distributions.

```python
""" Standardization, also known as Z-score normalization, 
    is a technique used to rescale the features of data to have a mean of 0 and a standard deviation of 1. 
    This is useful for algorithms that assume the data is normally distributed or that are sensitive to the scale of the data. """
```

```python
# Standardize numerical columns
standardized_data = sales[numerical_columns].apply(lambda x: (x - x.mean()) / x.std())
standardized_data = pd.concat([sales.drop(columns=numerical_columns), standardized_data], axis=1)

# Plot histograms before and after standardization
fig, axes = plt.subplots(2, 6, figsize=(20, 10))

# Original data histograms
for i, column in enumerate(numerical_columns):
    sns.histplot(sales[column], kde=True, ax=axes[0, i])
    axes[0, i].set_title(f'Original {column}')
    axes[0, i].set_xlabel(column)
    axes[0, i].set_ylabel('Frequency')

# Standardized data histograms
for i, column in enumerate(numerical_columns):
    sns.histplot(standardized_data[column], kde=True, ax=axes[1, i], color='red')
    axes[1, i].set_title(f'Standardized {column}')
    axes[1, i].set_xlabel(column)
    axes[1, i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Conversion of Categorical Data into Dummy Variables
#### â—	Objective: To transform categorical variables into a format that can be provided to ML algorithms.
#### â—	Steps:
	Discuss the need for converting categorical data into dummy variables (one-hot encoding).
	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
	Display a portion of the transformed dataset.

```python
""" Machine learning algorithms require numerical input, 
    but categorical data consists of labels that are not inherently numerical. 
    To use categorical data in machine learning models, we need to convert these labels into a numerical format. 
    One of the most common methods to achieve this is one-hot encoding. """
```

```python
""" One-hot encoding converts categorical variables into a series of binary (0 or 1) columns, 
    where each column represents a unique category. 
    This method ensures that the encoded data does not assume any ordinal relationship between categories, 
    making it suitable for algorithms that require numerical input without implying any order. """
```

```python
encoded_data = pd.get_dummies(sales, columns=categorical_columns)       # Apply one-hot encoding
encoded_data
```
